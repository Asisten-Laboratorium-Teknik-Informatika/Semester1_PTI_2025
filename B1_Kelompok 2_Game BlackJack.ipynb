```python
import random

def deal_card():
    cards = [2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10, 11]
    return random.choice(cards)

def hitungSkor(hand):
    if sum(hand) == 21 and len(hand) == 2:
        return 0   
    if 11 in hand and sum(hand) > 21:
        hand.remove(11)
        hand.append(1)
    return sum(hand)

def compare(skorPemain, skorBandar):
    if skorBandar == skorPemain:
        return "Seri, kamu kalah"
    elif skorBandar == 0:
        return "Kamu kalah, bandar blackjack!"
    elif skorPemain == 0:
        return "Menang dengan blackjack!"
    elif skorPemain > 21:
        return "Kamu kalah (bust)"
    elif skorBandar > 21:
        return "Kamu menang, bandar bust!"
    elif skorPemain > skorBandar:
        return "Kamu menang!"
    else:
        return "Kamu kalah"

def topUp(dwet):
    PW = "hahaha"
    maxTry = 3
    cobaPW = 0
    
    while cobaPW < maxTry:
        pwcoba = input("Masukkan password anda: ") 
        if pwcoba == PW:
            print("Login berhasil\n")
            break
        else:
            cobaPW += 1
            print("Password salah, coba lagi.")
    if cobaPW == maxTry: 
        print("Salah terus cik, sayonara..")
        return dwet
    
    jmlh = int(input("Masukkan jumlah top up : Rp."))
    if jmlh > 0:
        dwet += jmlh
        print(f"\nDeposit sebesar Rp.{jmlh} berhasil.")
    else:
        print("Masukan harus bilangan positif.")
    return dwet

def withdraw(dwet):
    jmlh = int(input("Masukkan jumlah withdraw : Rp."))
    if jmlh <= 0:
        print("Jumlah harus positif.")
    elif jmlh > dwet:
        print("Saldo tidak mencukupi untuk withdraw.")
    else:
        dwet -= jmlh
        print(f"Withdraw sebesar Rp.{jmlh} berhasil. Saldo tersisa Rp.{dwet}")
    return dwet

def main():
    print("blackjack77!\n!!!!PASTI GACOR KANG!!!!\n(please gamble responsibly)")
    dwet = 300000  
    history = []

    while True:  
        cois = int(input("\n1. masuk permainan\n2. histori permainan\n3. tips & tricks\n4. withdraw\n5. top up\n0. keluar\n"))
        
        if cois == 0:
            print("Terima kasih sudah bermain!")
            break
        if cois == 5:
            dwet = topUp(dwet)
            continue
        if cois == 4:
            dwet = withdraw(dwet)
            continue
        if cois == 2:
            print("\n=== Histori Permainan ===")
            if not history:
                print("Belum ada permainan.")
            else:
                for h in history:
                    print(h)
            continue
        if cois == 3:
            print("Tips & Tricks Blackjack:")
            print("1. Stand ketika kartu Anda 12-16 dan dealer memiliki 2-6")
            print("2. Hit ketika kartu Anda 12-16 dan dealer memiliki 7-As")
            print("3. Selalu pisahkan As dan 8")
            print("4. Gandakan taruhan pada 11 melawan 2-10 dealer")
            print("5. Pukul atau gandakan taruhan pada As-6")
            continue

        while cois == 1:
            print(f"\n||saldo skrg : Rp. {dwet}||")
            print("\n1. meja 1 (bet Rp. 500.000)\n2. meja 2 (bet Rp 250.000)\n3. meja 3 (bet Rp 100.000)\n0. keluar")    
            chois = False
            while not chois:
                meja = int(input("\nPilih meja :"))
                bet = 0
                if meja == 1:
                    bet = 500000
                elif meja == 2:
                    bet = 250000
                elif meja == 3:
                    bet = 100000
                elif meja == 0:
                    chois = True
                    cois = 0
                    break 
                else:
                    print("Meja tidak valid.")
                
                if dwet < bet:
                    print("\n||SALDO anda tidak mencukupi untuk bermain di meja ini, pilih meja lain.||\n")
                else:
                    chois = True
                    dwet -= bet
                    print(f"\n||Anda masuk ke meja {meja} dengan bet Rp. {bet}. Saldo tersisa Rp. {dwet}||\n")
            if meja == 0:
                continue
            kartuPemain = [deal_card(), deal_card()]
            kartuBandar = [deal_card(), deal_card()]

            game_over = False

            while not game_over:
                skorPemain = hitungSkor(kartuPemain)
                skorBandar = hitungSkor(kartuBandar)
                print(f"Kartu anda : {kartuPemain}, skor : {skorPemain}")
                print(f"Kartu bandar: {kartuBandar[0]}")

                if skorPemain == 0 or skorBandar == 0 or skorPemain > 21:
                    game_over = True
                else:
                    lanjut = input("Hit (h) or Stand (s)? ")
                    if lanjut == 'h':
                        kartuPemain.append(deal_card())
                    else:
                        game_over = True

            while skorBandar != 0 and skorBandar < 17:  
                kartuBandar.append(deal_card())
                skorBandar = hitungSkor(kartuBandar)
            
            print(f"\nKartu akhir pemain : {kartuPemain}, skor akhir : {skorPemain}")
            print(f"Kartu akhir bandar : {kartuBandar}, skor akhir : {skorBandar}")
            
            hasil = compare(skorPemain, skorBandar)
            print(hasil)

            if "menang" in hasil.lower():
                dwet += bet * 2

            print(f"Saldo sekarang: Rp. {dwet}")
            history.append(f"Meja {meja} | Hasil: {hasil} | Saldo: Rp.{dwet}")
            
            ulng = input("Ulang permainan? (y/n) ")
            if ulng != 'y':
                break

main()
```
